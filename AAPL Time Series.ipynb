# Setting up the Target for Machine Learning

```python
#Creating a new column showing the prices of the immediate next day

df['Tomorrow'] = df['Close'].shift(-1)
df.head()
```

```python
df['Target'] = (df['Tomorrow'] > df['Close']).astype(int)
df.head()
```

# Implementing a RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model_1 = RandomForestClassifier(n_estimators = 100,  min_samples_split=100, random_state=1)

train = df.iloc[:-50]
test = df.iloc[-50:]

predictors = ['Close', 'Volume', 'Open', 'High', 'Low']
model_1.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score

preds = model_1.predict(test[predictors])

preds= pd.Series(preds, index=test.index)
preds.head(25)
```

```python
precision_score(test['Target'], preds)
```

```python
combined = pd.concat([test['Target'], preds], axis=1)
combined.plot()
```
